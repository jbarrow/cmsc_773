```python
import pandas as pd
```

```python
i = 0
topic_matrix = []

with open('data/clpsych_doc_topics.txt', 'r') as fp:
    fp.readline()
    for i, line in enumerate(fp):
        # split the line into its components
        line = line.split()
        # get the post id
        key = line[1]
        # get the topic memberships
        topics = zip([int(t) for t in line[2::2]], [float(m) for m in line[3::2]])
        member = [0.,]*100
        for topic, prop in topics:
            member[topic] = prop
        topic_matrix.append([key] + member)
        if i % 100000 == 0: print i
topic_matrix = pd.DataFrame(topic_matrix)
```
